# 3. Display the percentage distribution of the data in each year using the pie chart. (2 marks) 

```python
percentage_distribution_year = df['year'].value_counts(normalize=True)*100
percentage_distribution_year

#--Calculating the percentage distribution of the data in each year
```

# The normalize=True argument is passed to calculate the relative frequencies of each unique value. This means it calculates the proportion of each year in the total dataset. This results in a Series where the indices are the unique years and the values are the corresponding relative frequencies (in decimal form).

```python
#---Plotting using pie chart

plt.figure(figsize=(10,5))
plt.pie(percentage_distribution_year, labels=percentage_distribution_year.index, autopct='%1.1f%%',startangle=140, shadow=True)
plt.title('percentage distribution of the data in each year')
plt.show()
```

# 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings. (2 marks) 

```python
#---Plotting 'Distribution of the variable priceperlb' using displot 

sns.displot(data=df, x=df['priceperlb'], kind='kde')
plt.title('Distribution of the variable priceperlb')
plt.xlabel('priceperlb')
plt.ylabel('Frequency')
plt.show()
```

```python
# - Findings:

#-- The distribution is right-skewed, indicating that there are fewer data points with higher prices per pound.
#-- The majority of the prices per pound are concentrated on the lower end of the scale.
#-- There are a few instances of relatively higher prices per pound, but they are not as common.

```

# 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings. (2 marks)

```python
#---Plotting the relationship between the variables 'numcol' and 'prodval' using scatterplot

plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x=df['numcol'], y=df['prodvalue'])
plt.title('relationship between numcols and prodvalue')
plt.show()
```

```python
#---Findings

# There appears to be a positive correlation between the number of colonies ('numcol') and the production value ('prodvalue'). As the number of colonies increases, the production value tends to increase as well.
# However, the relationship is not perfectly linear. There is some variation in the production value for each number of colonies, indicating that other factors might be influencing production value as well.
# There are a few instances where the number of colonies is relatively high, but the production value is relatively low, suggesting that there might be cases with lower efficiency or other influencing factors.
```

#  6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings. (2 marks)

```python
#---Plotting the relationship between categorical variable 'year' and a numerical variable 'prodvalue' using boxplot

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x=df['year'], y=df['prodvalue'])
plt.title('Box Plot of prodvalue by Year')
plt.xlabel('year')
plt.ylabel('prodvalue')
plt.show()
```

```python
#---Findings

# There seems to be a noticeable variation in production values across different years.
# The median production value appears to increase from 1998 to later years (if there's a trend present).
# There might be some outliers in certain years, indicating unusually high or low production values in those years.
```

# 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue') (2 marks)

```python
columns = ['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']
columns

#---Columns that needs to be correlated
```

```python
#--Plotting a heatmap of the correlation matrix

plt.figure(figsize=(10,5))
sns.heatmap(data=df[columns].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# --Inferences:

#-- The color scale in the heatmap indicates the strength and direction of correlations. Positive correlations are shown in warmer colors (red tones), while negative correlations are shown in cooler colors (blue tones).

#-- Strong positive correlations are observed between 'totalprod' and 'numcol', as well as between 'prodvalue' and 'totalprod'. This suggests that higher numbers of colonies and higher total production are associated with higher production values.

#-- There is a moderate negative correlation between 'priceperlb' and 'prodvalue'. This indicates that as the price per pound increases, the production value tends to decrease.

#-- Other correlations appear to be relatively weak, suggesting that the variables are less strongly related.
```
